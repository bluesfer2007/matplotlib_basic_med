```python
import matplotlib.pyplot as plt
import numpy as np

def draw_meter(percentage, ax=None):
    if ax is None:
        ax = plt.subplot(111, polar=True)
    else:
        ax = plt.subplot(111, polar=True, ax=ax)

    # Ocultar el marco y los ticks
    ax.axis('off')

    # Dibujar el fondo del tac√≥metro
    ax.barh(1, np.radians(180), color='green', left=np.radians(180))

    # Calcular el √°ngulo para el porcentaje
    angle = np.radians(percentage / 100 * 180 + 180)

    # Dibujar la aguja
    ax.barh(1, np.radians(0.5), color='black', left=angle)

    # Ajustes adicionales
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)

    plt.show()

# Llamar a la funci√≥n con el 75% de avance
draw_meter(75)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Wedge, Arc

def draw_horizontal_meter(percentage):
    fig, ax = plt.subplots(subplot_kw={'aspect': 'equal'})

    # Dibujar el arco del tac√≥metro
    arc = Arc((0.5, 0.5), 1, 1, angle=0, theta1=0, theta2=180, edgecolor='black', lw=2)
    ax.add_patch(arc)

    # Calcular la posici√≥n de la aguja en base al porcentaje
    needle_angle = percentage / 100 * 180
    needle_x = 0.5 + 0.5 * np.cos(np.radians(180 - needle_angle))
    needle_y = 0.5 + 0.5 * np.sin(np.radians(180 - needle_angle))

    # Dibujar la aguja
    ax.plot([0.5, needle_x], [0.5, needle_y], color='red', lw=2)

    # Ajustar l√≠mites para mantener la proporci√≥n
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 0.5)

    # Ocultar los ejes
    ax.axis('off')

    plt.show()

# Llamar a la funci√≥n con el 75% de avance
draw_horizontal_meter(75)
```
