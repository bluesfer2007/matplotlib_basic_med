```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_excel('DataTurbinasFly.xlsx')
```

```python
hreal=df.horas.groupby(df.Year).agg('sum')
hplan=df.proyec.groupby(df.Year).agg('sum')
```

## GRAFICO BASICO

```python
extra_year=[2024,2025, 2026]
val_adic=[6500000, 7000000,5500000]

plt.plot(hreal, c='r')
plt.plot(hplan)
plt.plot(extra_year,val_adic)
plt.title('VALORES HORAS REAL', c='#0B60B0')
plt.grid(True, linestyle=':', c='b')
plt.xlabel('years values', c='#0B60B0')
plt.ylabel('values', c='#0B60B0')
plt.text(2024, 6000000, 'this data \n dummy', c='g')


plt.show()
```

## OBJECT FIGURE AND AX

```python
years=list(set(df.Year))
fig, ((ax1, ax2), (ax3, ax4))=plt.subplots(2,2, figsize=[12,6])
ax1.plot(hreal, 'ro')
ax4.plot(hplan)
ax1.set_title('Title AX1')
ax4.set_title('title AX4')
fig.suptitle('title figure general')
ax4.grid(True)
ax1.text(2022, 6000000, 'any text to describe')
#ax2.text(0.5, 0.6, 'Text to name anything')
#ax3.bar(years,hreal)
ax3.set_xticks(years)
bar=ax3.bar(years,hreal)
bar[3].set_color('#561C24')
ax3.text(2023, 4000000, 'test \n text', c='w')
ax2.plot(hreal, label='real hour')
ax2.plot(hplan, label='plan hour')
ax2.set_title('Real vs Plan')
leg=ax2.legend( frameon=False)

#plt.show()
plt.savefig('name_figure_axes', dpi=1000)
```
